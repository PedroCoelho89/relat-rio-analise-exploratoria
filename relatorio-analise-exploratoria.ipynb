```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("Bibliotecas importadas com sucesso!")
```

```python
orders = pd.read_csv('olist_orders_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
geolocation = pd.read_csv('olist_geolocation_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
category_translation = pd.read_csv('product_category_name_translation.csv')


print("Todos os arquivos foram carregados com sucesso!")
```

```python
# A "chave" que conecta as duas é a coluna 'order_id
df = pd.merge(orders, items, on='order_id', how='left')

# A chave aqui é 'customer_id'
df = pd.merge(df, customers, on='customer_id', how='left')

# A chave para conectar é 'order_id' novamente.
df = pd.merge(df, payments, on='order_id', how='left')

# A chave que conecta o produto ao item do pedido é 'product_id'.
df = pd.merge(df, products, on='product_id', how='left')

# A chave que conecta o item ao vendedor é 'seller_id'
df = pd.merge(df, sellers, on='seller_id', how='left')

# Adicionando a tradução das categorias. A chave é 'product_category_name'
df = pd.merge(df, category_translation, on='product_category_name', how='left')

# Juntando o nosso dataset 'df' com o de 'reviews'
df_com_reviews = pd.merge(df, reviews, on='order_id', how='left')
df_com_reviews.head()

print("Todas as tabelas foram juntadas numa só!")
print("\nSeu novo dataset unificado tem o seguinte formato:")
print("Número de Linhas:", df.shape[0])
print("Número de Colunas:", df.shape[1])

print("\nVeja as 5 primeiras linhas do seu dataset geral:")
df.head()
```

```python
# Apenas para garantir que estamos trabalhando com o data correto
df = df_com_reviews.copy()
```

```python
# Mostra a contagem de valores nulos para cada coluna
print(df.isnull().sum())
```

```python
# Remove as linhas onde qualquer uma dessas colunas de data esteja nula
colunas_criticas = ['order_delivered_customer_date', 'order_purchase_timestamp', 'order_estimated_delivery_date']
df.dropna(subset=colunas_criticas, inplace=True)

print("Linhas com datas críticas faltando foram removidas.")
print("Verificando novamente os nulos nessas colunas:")
print(df[colunas_criticas].isnull().sum())
```

```python
# Lista de colunas que devem ser do tipo data
colunas_de_data = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'review_creation_date',
    'review_answer_timestamp'
]

# Loop para converter cada coluna para datetime
for coluna in colunas_de_data:
    df[coluna] = pd.to_datetime(df[coluna])

print("Colunas de data convertidas para o formato datetime com sucesso!")
df.info()
```

```python
# Lógica: Data que o cliente recebeu - Data que o pedido foi feito
df['tempo_de_entrega'] = (df['order_delivered_customer_date'] - df['order_purchase_timestamp']).dt.days


# Lógica: Data que o cliente recebeu - Data que era estimada para a entrega
df['atraso_na_entrega'] = (df['order_delivered_customer_date'] - df['order_estimated_delivery_date']).dt.days


# Uma coluna de texto que facilita a análise: 'Atrasado' ou 'No Prazo'
# Se o 'atraso_na_entrega' for maior que 0, então atrasou.
import numpy as np
df['status_atraso'] = np.where(df['atraso_na_entrega'] > 0, 'Atrasado', 'No Prazo')


print("Novas colunas de features criadas com sucesso!")
# Vamos visualizar as novas colunas junto com as datas e a nota da avaliação para conferir
colunas_para_verificar = [
    'order_purchase_timestamp',
    'order_estimated_delivery_date',
    'order_delivered_customer_date',
    'tempo_de_entrega',
    'atraso_na_entrega',
    'status_atraso',
    'review_score'
]

df[colunas_para_verificar].sample(5)
```

```python
# Salva o dataframe final, com todas as limpezas e novas features, em um novo arquivo CSV.
# O index=False impede o pandas de salvar o índice do dataframe como uma coluna extra.
df.to_csv('dataset_final_tratado.csv', index=False)

print("A fase de preparação de dados com Python está completa.")
print("O arquivo 'dataset_final_tratado.csv' foi salvo no seu ambiente colab.")
print("Ele está pronto para a segunda fase do projeto: análise e visualização com R.")
```
